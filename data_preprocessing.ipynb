```python
import requests
import configparser
import json
import pandas as pd
```

```python
# Read Conf
config = configparser.ConfigParser()
config.read('conf.ini')
alphavantage_api = config['SETTINGS']['alphavantage_api']
fmp_api = config['SETTINGS']['fmp_api']
```

```python

market_list = ['SPY','QQQ','TLT','IWM','^VIX','^VVIX','^TNX']

def get_historical_data(time,fmp_api):
    
    for item in market_list:
        url = 'https://financialmodelingprep.com/api/v3/historical-price-full/%s?timeseries=%s&apikey=%s' % (item,time,fmp_api)
        #print(url)
        resp = requests.get(url=url)
        data = resp.json()
        hist_df = pd.json_normalize(data['historical'])
        
        hist_df = hist_df.add_prefix(item + '_')
        
        if '^' in item:
            pass
        else:
            hist_df[item + '_vol_changePercent'] = hist_df[item + '_volume'].pct_change(periods=-1)
            
        
        #print(hist_df.shape)
        #print(hist_df.columns)

        if item == 'SPY':
            df = hist_df[['SPY_date','SPY_changePercent','SPY_vol_changePercent', 'SPY_close']]
            df=df.rename(columns = {'SPY_date':'date'})
        elif '^' in item:
            tmp = hist_df[[item+'_date',item+'_changePercent']]
            df = df.merge(tmp, how='left', left_on='date', right_on=item+'_date')
            df = df.drop(columns=[item+'_date'])
        else:
            tmp = hist_df[[item+'_date',item+'_changePercent',item+'_vol_changePercent']]
            df = df.merge(tmp, how='left', left_on='date', right_on=item+'_date')
            df = df.drop(columns=[item+'_date'])
    
    return df


        #df = 
        
    #spy_hist = pd.json_normalize(data['historical'])
    #print(spy_hist.shape)
    

df = get_historical_data('1500',fmp_api)



df.head()
```
