```python
import requests
import configparser
import json
import pandas as pd
```

```python
# Read Conf
config = configparser.ConfigParser()
config.read('conf.ini')
alphavantage_api = config['SETTINGS']['alphavantage_api']
fmp_api = config['SETTINGS']['fmp_api']
```

```python
# read fata from fmp
url = 'https://financialmodelingprep.com/api/v3/historical-price-full/SPY?timeseries=1500&apikey=' 
resp = requests.get(url=url+fmp_api)
data = resp.json()
spy_hist = pd.json_normalize(data['historical'])
print(spy_hist.shape)
```

```python
url = 'https://financialmodelingprep.com/api/v3/historical-price-full/TLT?timeseries=1500&apikey=' 
resp = requests.get(url=url+fmp_api)
data = resp.json()
tlt_hist = pd.json_normalize(data['historical'])
print(tlt_hist.shape)
```
